#**Libraries :**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

#**DATASET LOADING :**

```python
Data = pd.read_csv('/content/college_student_placement_dataset.csv')
df = Data
```

#**DATA SPLIT :**

```python
df.head()
```

```python
df["Placement"] = df["Placement"].map({"Yes":1,"No":0})
df["Internship_Experience"] = df["Internship_Experience"].map({"Yes":1,"No":0})
df
```

```python
df.dtypes
```

```python
x=df.drop(["College_ID","Placement"],axis=1)
y=df["Placement"]
x
```

```python
y
```

```python
print(x.shape)
print(y.shape)
```

```python
df["Placement"].value_counts()
```

#**TRAIN TEST SPLIT :**

```python
X_train,x_test,Y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

#**RANDOM FOREST MODEL :**

```python
Student= RandomForestClassifier(n_estimators=100,random_state=42)
Student.fit(X_train,Y_train)
```

#**CREATE PREDICTION :**

```python
y_pred = Student.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
print(f"placement accuracy:{acc}")
```

```python
from enum import CONFORM
CFM=confusion_matrix(y_test,y_pred)
print("placement confusion matrix:",CFM)
```

#**CLASSIFICATION REPORT :**

```python
CFR = classification_report(y_test,y_pred)
print(CFR)
```

#**VISUALIZATION USING HEATMAPS :**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

label=["Not Placing","Placing"]
plt.figure(figsize=(8,6))

sns.heatmap(cm,annot=True,fmt="d",xticklabels=label,yticklabels=label,cmap="Greens")
plt.title("CONFUSION MATRIX - HEATMAP")
plt.xlabel("Predicted Condition")
plt.ylabel("Actual Condition")
plt.show()
```
